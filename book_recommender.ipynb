```python
ratings_with_books=ratings.merge(books,on="ISBN")
ratings_with_books
```

```python
num_rating_df=ratings_with_books.groupby('Book-Title').count()['Book-Rating']
num_rating_df
```

```python
num_rating_df=ratings_with_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
ratings_with_books['Book-Rating'] = pd.to_numeric(ratings_with_books['Book-Rating'], errors='coerce')
avg_rating_df = ratings_with_books.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

## Collaborative Filtering

```python
X=ratings_with_books.groupby('User-ID').count()['Book-Rating'] > 200
users_ratings_submitted=X[X].index
```

```python
filtered_rating=ratings_with_books[ratings_with_books['User-ID'].isin(users_ratings_submitted)]
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
# filtered_rating
```

```python
final_ratings.drop_duplicates()
```

```python
pivot=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot
```

```python
pivot.fillna(0,inplace=True)
```

```python
pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score=cosine_similarity(pivot)
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pivot.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pivot.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```
